```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
```

```python
data=pd.read_csv('Fertilizer Prediction.csv')
```

```python
cat_dat=data.select_dtypes(include='object')
```

```python
y=cat_dat['Fertilizer Name']
cat_dat.drop('Fertilizer Name',axis=1,inplace=True)
```

```python
cat_ohe=pd.get_dummies(cat_dat)
cat_ohe.head()
```

```python
numerical=data.select_dtypes(include='int')
numerical.head()
```

```python
col=0
fig,ax=plt.subplots(3,2,figsize=(12,8))
for i in numerical.columns.tolist():
    sns.boxplot(data=numerical,x=i,ax=ax.flat[col])
    col+=1
plt.tight_layout()  
```

```python
num_nor=(numerical-numerical.mean())/(numerical.std())
num_nor
```

```python
X=pd.concat([cat_ohe,num_nor],axis=1)
X.head()
```

```python
d={}
for i,j in enumerate(y.unique().tolist()):
    d[j]=i
d
```

```python
y=data['Fertilizer Name'].map(d)
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
y_pred=log.fit(X_train,y_train).predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rand=RandomForestClassifier()
y_rand=rand.fit(X_train,y_train).predict(X_test)
```

```python
accuracy_score(y_rand,y_test)
```
